```python
import requests # HTTP 요청을 보내기 위한 라이브러리 임포트

def get_headers(url):
    """지정된 URL에서 서버 헤더 정보를 가져옵니다."""
    try:
        response = requests.get(url, timeout=10) # 10초 타임아웃으로 GET 요청을 보냅니다.
        return response.headers # 응답에서 헤더 정보를 반환합니다.
    except requests.exceptions.RequestException as e:
        # 요청 중 오류가 발생하면 오류 메시지를 반환합니다.
        return f"{url}에서 헤더 정보를 가져오는 중 오류 발생: {e}"

# 비교할 웹사이트 URL을 정의합니다.
naver_url = 'https://www.naver.com'
google_url = 'https://www.google.com'

print(f"--- {naver_url} 의 헤더 정보 ---")
naver_headers = get_headers(naver_url) # 네이버 헤더 정보를 가져옵니다.
if isinstance(naver_headers, dict):
    for key, value in naver_headers.items():
        print(f"{key}: {value}") # 각 헤더 키와 값을 출력합니다.
else:
    print(naver_headers) # 오류 메시지가 있으면 출력합니다.

print(f"\n--- {google_url} 의 헤더 정보 ---")
google_headers = get_headers(google_url) # 구글 헤더 정보를 가져옵니다.
if isinstance(google_headers, dict):
    for key, value in google_headers.items():
        print(f"{key}: {value}") # 각 헤더 키와 값을 출력합니다.
else:
    print(google_headers) # 오류 메시지가 있으면 출력합니다.
```

```python
import pandas as pd
import gradio as gr
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import io

# 한글 폰트 설정 (Colab 환경)
# 런타임 재시작이 필요할 수 있습니다.
!apt-get update -qq
!apt-get install fonts-nanum* -qq

font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# CSV 파일 경로
csv_file_path = '/content/서울특별시 따릉이 10월,11월,12월.csv'

# 데이터를 한 번 전역적으로 로드합니다.
try:
    df_따릉이 = pd.read_csv(csv_file_path, encoding='cp949')
except FileNotFoundError:
    print(f"오류: 파일 '{csv_file_path}'를 찾을 수 없습니다.")
    df_따릉이 = pd.DataFrame() # 파일이 없으면 빈 DataFrame을 제공하여 오류 방지
except UnicodeDecodeError:
    print(f"오류: 파일 '{csv_file_path}'의 인코딩을 감지할 수 없습니다. 'utf-8'을 시도합니다.")
    df_따릉이 = pd.read_csv(csv_file_path, encoding='utf-8')

def get_column_names():
    """로드된 DataFrame에서 컬럼 이름 목록을 반환합니다."""
    if not df_따릉이.empty:
        return df_따릉이.columns.tolist()
    return []

def get_dataframe_info(df):
    """DataFrame의 info() 결과를 텍스트로 반환합니다."""
    buffer = io.StringIO()
    df.info(buf=buffer)
    return buffer.getvalue()

def perform_analysis(column_name):
    """
    선택된 컬럼에 대해 기본 통계 분석을 수행하고 그래프를 생성합니다.
    """
    if df_따릉이.empty or column_name is None:
        return pd.DataFrame({"Message": ["데이터를 로드하거나 컬럼을 선택해주세요."]}, index=['Info']), None, None

    # 기술 통계
    if pd.api.types.is_numeric_dtype(df_따릉이[column_name]):
        stats = df_따릉이[column_name].describe().to_frame().T
    else:
        stats_series = df_따릉이[column_name].value_counts()
        stats = stats_series.to_frame()
        stats.columns = ['Count'] # 컬럼 이름 변경
        stats.index.name = 'Value' # 인덱스 이름 설정

    # 숫자형 컬럼에 대해서만 그래프 생성
    hist_fig = None
    boxplot_fig = None
    if pd.api.types.is_numeric_dtype(df_따릉이[column_name]):
        # 히스토그램
        plt.figure(figsize=(10, 5))
        sns.histplot(df_따릉이[column_name].dropna(), kde=True)
        plt.title(f'{column_name} 히스토그램')
        plt.xlabel(column_name)
        plt.ylabel('빈도')
        hist_fig = plt.gcf()
        plt.close(hist_fig) # Gradio 외부 표시 방지를 위해 닫기

        # 상자 그림 (Box Plot)
        plt.figure(figsize=(10, 5))
        sns.boxplot(y=df_따릉이[column_name].dropna())
        plt.title(f'{column_name} 상자 그림')
        plt.ylabel(column_name)
        boxplot_fig = plt.gcf()
        plt.close(boxplot_fig) # Gradio 외부 표시 방지를 위해 닫기
    else:
        # 비숫자형 컬럼의 경우, 값 분포를 막대 그래프로 표시
        # 고유값이 너무 많으면 그래프를 그리지 않습니다.
        if len(df_따릉이[column_name].unique()) < 20 and not df_따릉이[column_name].empty:
            plt.figure(figsize=(10, 5))
            sns.countplot(y=df_따릉이[column_name].dropna(), order=df_따릉이[column_name].value_counts().index)
            plt.title(f'{column_name} 값 분포 (막대 그래프)')
            plt.xlabel('빈도')
            plt.ylabel(column_name)
            hist_fig = plt.gcf()
            plt.close(hist_fig)

    return stats, hist_fig, boxplot_fig

# 초기 컬럼 선택 및 분석 결과 설정
initial_choices = get_column_names()
# '대여건수' 또는 '반납건수'와 같은 숫자형 컬럼을 초기 선택 값으로 설정합니다.
# 만약 '대여건수'가 없으면 첫 번째 숫자형 컬럼을 선택하고, 없으면 첫 번째 컬럼을 선택합니다.
numeric_cols = [col for col in initial_choices if pd.api.types.is_numeric_dtype(df_따릉이[col])]
if '대여건수' in numeric_cols:
    initial_column_selected = '대여건수'
elif numeric_cols:
    initial_column_selected = numeric_cols[0]
else:
    initial_column_selected = initial_choices[0] if initial_choices else None

initial_stats, initial_hist, initial_boxplot = perform_analysis(initial_column_selected)

# Gradio Blocks 인터페이스 생성
with gr.Blocks() as demo:
    gr.Markdown("# 서울특별시 따릉이 데이터 다중 화면 분석")
    gr.Markdown(f"파일 `{csv_file_path}`를 로드하여 여러 화면에서 데이터를 탐색하고 분석합니다.")

    with gr.Tabs():
        with gr.TabItem("데이터 개요"): # 첫 번째 탭: 데이터 개요
            gr.Markdown("### 데이터프레임 미리보기 (상위 5행)")
            gr.DataFrame(value=df_따릉이.head())

            gr.Markdown("### 데이터프레임 정보 (info())")
            gr.Code(value=get_dataframe_info(df_따릉이), lines=10) # language='text' 제거

            gr.Markdown("### 기술 통계 요약 (describe())")
            gr.DataFrame(value=df_따릉이.describe())

        with gr.TabItem("컬럼별 상세 분석"): # 두 번째 탭: 컬럼별 상세 분석
            gr.Markdown("### 컬럼 선택 및 상세 분석")
            with gr.Row():
                column_dropdown = gr.Dropdown(
                    label="분석할 컬럼 선택",
                    choices=initial_choices,
                    value=initial_column_selected, # 초기 선택 값 설정
                    interactive=True
                )

            with gr.Row():
                stats_output = gr.DataFrame(label="기술 통계", value=initial_stats) # 초기 통계 데이터 표시

            with gr.Row():
                histogram_output = gr.Plot(label="히스토그램 / 값 분포", value=initial_hist) # 초기 히스토그램 표시
                boxplot_output = gr.Plot(label="상자 그림", value=initial_boxplot) # 초기 상자 그림 표시

            # 드롭다운 변경 시 분석 함수 호출
            column_dropdown.change(
                fn=perform_analysis,
                inputs=column_dropdown,
                outputs=[stats_output, histogram_output, boxplot_output]
            )

demo.launch(share=True)
```
